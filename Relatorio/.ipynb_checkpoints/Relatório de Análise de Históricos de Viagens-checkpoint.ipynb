# Relatório de Análise de Históricos de Viagens

```python
import matplotlib.pyplot as plt
from pandas import DataFrame, read_csv, ExcelFile, ExcelWriter
import pandas as pd
from datetime import datetime
import pytz
```

# Realiza o cálculo que determina a situação do motorista. Sendo elas, atrasado, adiantado ou sem atraso

```python
def df_hours(hora_prevista, hora_entreposto):
    
    dif = hora_prevista - hora_entreposto
    dif_in_s = dif.total_seconds()
    dias = divmod(dif_in_s,86400)
    horas = divmod(dias[1],3600)
    minutos = divmod(horas[1],60)
    segundos = divmod(minutos[1],1)
    
    if dias[0] >= 1 or (horas[0] >= 1 and minutos[0] >= 1):
         return 'ATRASADO'
    elif dias[0] <= -1 or (horas[0] >= 1 and minutos[0] >= 1):
        return 'ADIANTADO'
    else:
        return 'SEM ATRASO'
```

# Faz a leitura do arquivo .xls e retorna os vetores contendo motoristas atrasados, adiantados e sem atraso

```python
def reader():
    file = r'data/viagens.xls'
    df = pd.read_excel(file)
    chegada_prevista, chegada_entreposto =  df['Chegada Prevista'], df['Chegada ao Entreposto']
    lenght = len(chegada_prevista)
    
    atrasado = list()
    sem_atraso = list()
    adiantado = list()
    
    timezone = pytz.timezone('America/Sao_Paulo')
    
    for i in range(lenght):

        if pd.isna(chegada_prevista[i]):
            continue
        else:
            time_prevista = datetime.strptime(chegada_prevista[i], '%d/%m/%Y %H:%M:%S')
            timezone_date_obj_prevista = timezone.localize(time_prevista)

        
        if pd.isna(chegada_entreposto[i]):
            continue
        else:
            time_entreposto = datetime.strptime(chegada_entreposto[i],'%d/%m/%Y %H:%M:%S')
            timezone_date_obj_entreposto = timezone.localize(time_entreposto)
        
        aux = df_hours(timezone_date_obj_entreposto, timezone_date_obj_prevista)
        if aux == 'ATRASADO':
            atrasado.append(aux)
        elif aux == 'ADIANTADO':
            adiantado.append(aux)
        else:
            sem_atraso.append(aux)
            
    return atrasado, sem_atraso, adiantado
```

# Gera o gráfico de pizza para motoristas sem atraso e atrasados

```python
atrasado, sem_atraso, adiantado = reader()

labels = 'ATRASADO', 'SEM ATRASO'
sizes = [len(atrasado), len(sem_atraso)]
explode = (0,0.1)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.title('Viagens atrasadas')
plt.show()
```

# Gera o gráfico de pizza para motoristas sem atraso e adiantados

```python
labels = 'ADIANTADO', 'SEM ATRASO'
sizes = [len(adiantado), len(sem_atraso)]
explode = (0.1,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.title('Viagens adiantadas')
plt.show()
```

# Gera o gráfico com todas as situações possíveis dos motoristas

```python
labels = 'ATRASADO', 'SEM ATRASO', 'ADIANTADO'
sizes = [len(atrasado), len(sem_atraso), len(adiantado)]
explode = (0.1,0,0.1)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.title('Viagens')
plt.show()
```
