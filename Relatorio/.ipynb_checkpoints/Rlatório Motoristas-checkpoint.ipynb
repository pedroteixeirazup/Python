```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def qtd_motoristas(motoristas):
    motoristas_set = set()
    
    for i in motoristas:
        motoristas_set.add(i)

    m = list(motoristas)
    motoristas_arr = list(motoristas_set)
    dic = {}
    
    for i in motoristas_arr:
        dic[i] = m.count(i)

    return len(motoristas_set), sorted(dic.items(), key =lambda x: x[1], reverse=True)[:10]
```

```python
def qtd_veiculos(veiculos):
    veiculos_set = set()
    
    for i in veiculos:
        veiculos_set.add(i)
    
    v = list(veiculos)
    veiculos_arr = list(veiculos_set)
    dic = {}
    
    for i in veiculos_arr:
        dic[i] = v.count(i)
    
    return len(veiculos_set), sorted(dic.items(), key =lambda x: x[1], reverse=True)[:10]
    
```
